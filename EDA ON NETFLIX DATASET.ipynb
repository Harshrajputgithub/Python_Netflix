```python
import pandas as pd
data=pd.read_csv('netflix_dataset.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data['director'].fillna('No Data',inplace=True)
data['cast'].fillna('No Data',inplace=True)
```

```python
data.isna().sum()
```

```python
import statistics
country=statistics.mode(data['country'])
date_added=statistics.mode(data['date_added'])
rating=statistics.mode(data['rating'])
```

```python
data['country'].fillna(country,inplace=True)
data['date_added'].fillna(date_added,inplace=True)
data['rating'].fillna(rating,inplace=True)
```

```python
data.isna().sum()
```

```python
total=data['type'].count()
```

```python
tv_show=data[data['type']=='TV Show'].shape[0]
movie=data[data['type']=='Movie'].shape[0]
import numpy as np
percent_tv=np.true_divide(tv_show,total)*100
percent_movie=np.true_divide(movie,total)*100
```

```python
import matplotlib.pyplot as plt
size=[percent_tv,percent_movie]
labels=['percent_tv','percent_movie']
plt.pie(size,labels=labels,autopct='%.2f%%')
plt.title('percentage of tv_show and movies')
plt.show()
```

```python
data.head()
```

```python
country_data=data.groupby('country')['type'].count().sort_values(ascending=False)[:10]
index=country_data.index
value=country_data.values

plt.bar(index,value,color=('blue','red','orange'))
plt.xticks(rotation=45)
plt.show()

top_country=country_data.idxmax()
top_content=country_data.max()
output=(top_country,top_content)
print(output)
```

```python

data['year_added']=pd.to_datetime(data['date_added']).dt.year
a=data['year_added'].value_counts().index
b=data['year_added'].value_counts().values
plt.bar(a,b)
plt.title('content added over the year')
plt.show()
```

```python
movies_data = data[data['type'] == 'Movie']

movies_data['duration'] = movies_data['duration'].str.extract('(\d+)').astype(float)

longest_movie = movies_data[movies_data['duration'] == movies_data['duration'].max()]

movie_title = longest_movie['title'].values[0]
release_year = longest_movie['release_year'].values[0]

output=((movie_title), (release_year))
print(output)
```
